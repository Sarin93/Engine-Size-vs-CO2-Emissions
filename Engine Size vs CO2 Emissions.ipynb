```python
# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Read the data from CSV file
data = pd.read_csv("FuelConsumption.csv")
```

```python
# Step 3: Data Preprocessing
data = data[['ENGINESIZE', 'CO2EMISSIONS']]  # Selecting relevant columns
x = data[['ENGINESIZE']]  # Independent variable
y = data['CO2EMISSIONS']  # Dependent variable
```

```python
# Step 4: Split the data into training and testing sets
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
```

```python
# Step 5: Model Training
Simple_linReg = LinearRegression()
Simple_linReg.fit(xtrain, ytrain)
```

```python
# Step 6: Model Testing and Evaluation
ypred = Simple_linReg.predict(xtest)
r2 = r2_score(ytest, ypred)
```

```python
# Step 7: Visualization
plt.figure(figsize=(10, 6))
sns.scatterplot(x=xtest.values.ravel(), y=ytest, label="Actual Data")
sns.lineplot(x=xtest.values.ravel(), y=ypred, color="red", label="Predicted Data")
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emissions")
plt.title("Engine Size vs CO2 Emissions")
plt.legend()
plt.show()
```
